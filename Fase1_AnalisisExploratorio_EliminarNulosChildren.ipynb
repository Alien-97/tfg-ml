```
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

# En la columna Country, se sustituyen los valores nulos por la categor√≠a 'Unknown'
with open('/content/drive/My Drive/TFG_AlienEmbarecRiadi/hotel_bookings_modified.csv') as csv_file:
  df = pd.read_csv(csv_file, delimiter=',')
  df['market_segment'] = np.where(df['market_segment'] == 'Undefined', 'Unknown', df['market_segment'])
  df['distribution_channel'] = np.where(df['distribution_channel'] == 'Undefined','Unknown', df['distribution_channel'])
  print(df.info())
  df['children'] = df['children'].astype(np.int64)
  print(df.info())
  #df[df[0].notnull()] = df[df[0].notnull()].astype(int)
  df['agent'] = pd.array(df['agent'], dtype=pd.Int64Dtype())
  print(df['agent'])
  df.to_csv("/content/drive/My Drive/TFG_AlienEmbarecRiadi/hotel_bookings_modified.csv", index=False)
  
  #print(df.head(50))
```
